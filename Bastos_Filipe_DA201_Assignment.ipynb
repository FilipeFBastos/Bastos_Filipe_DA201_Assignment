```python
# Create a new DataFrame containing only the text.
tweets_text = tweets[['tweet_full_text', 'tweet_entities', 'tweet_entities_hashtags', 'tweet_metadata', 'tweet_lang']]

# View the DataFrame.
tweets_text
```

```python
# Loop through the messages, and create a list of values containing the # symbol.
tags = []

for y in [x.split(' ') for x in tweets['tweet_full_text'].values]:
    for z in y:
        if '#' in z:
            # Change to lowercase.
            tags.append(z.lower())
```

```python
# Display the first 30 records.
tags_s = pd.Series(tags).value_counts()

tags_s[:30]
```

```python
# Convert the series to a DataFrame in preparation for visualisation.
data = pd.DataFrame(tags_s).reset_index()

# Rename the columns.
data = data.rename(columns={'index': 'word', 0 : 'count'})
data.dtypes
```

### Investigate the main cencerns posed by the NHS. 

```python
# Prepare your workstation.
# Load the appointments_regional.csv file.
ar = pd.read_csv('appointments_regional.csv')

# View the DataFrame.
ar.head()
```

```python
# Print the min and max dates.
print(ar['appointment_month'].min())
print(ar['appointment_month'].max())
```

```python
# Filter the data set to only look at data from 2021-08 onwards.
ar = ar[ar['appointment_month'] >= '2021-08']
ar.head()
```

**Question 1:** Should the NHS start looking at increasing staff levels? 

```python
# Create an aggregated data set to review the different features.
ar_agg = ar.groupby([ar['appointment_month'], ar['hcp_type'], ar['appointment_status'], ar['appointment_mode'], 
        ar['time_between_book_and_appointment']])[['count_of_appointments']].agg(sum).reset_index()

# View the DataFrame.
ar_agg 
```

```python
# Determine the total number of appointments per month.
ar_df = ar.groupby([ar['appointment_month']])[['count_of_appointments']].agg(sum).reset_index()

# Add a new column to indicate the average utilisation of services.
# Monthly aggregate / 30 to get to a daily value.
ar_df['utilisation'] = ar_df['count_of_appointments']/30
ar_df['utilisation'] = ar_df['utilisation'].round(1)

# View the DataFrame.
ar_df
```

```python
# Plot sum of count of monthly visits.
# Convert the appointment_month to string data type for ease of visualisation.
ar_agg.astype({'appointment_month' : 'str'})

# Create a lineplot with Seaborn.
sns.lineplot(x="appointment_month", y="count_of_appointments", data=ar_agg, ci=None)
```

```python
# Plot monthly capacity utilisation.
ar_df.astype({'appointment_month' : 'str'})

# Create a lineplot.
sns.lineplot(x="appointment_month", y="utilisation", data=ar_df)
```

**Question 2:** How do the healthcare professional types differ over time?

```python
# Determine the total number of appointments per month.
ar_hcp = ar.groupby([ar['appointment_month'], ar['hcp_type']])[['hcp_type']].value_counts().reset_index()

# Create a line plot to answer the question.
sns.lineplot(x="appointment_month", y=0, hue ='hcp_type', data=ar_hcp, ci=None)
```

**Question 3:** Are there significant changes in whether or not visits are attended?

```python
# Determine the total number of appointments per month.
ar_hcp = ar.groupby([ar['appointment_month'], ar['appointment_status']])[['appointment_status']].value_counts().reset_index()

# Create a line plot to answer the question.
sns.lineplot(x="appointment_month", y=0, hue ='appointment_status', data=ar_hcp, ci=None)
```

**Question 4:** Are there changes in terms of appointment type and the busiest months?

```python
# Determine the total number of appointments per month.
ar_hcp = ar.groupby([ar['appointment_month'], ar['appointment_mode']])[['appointment_mode']].value_counts().reset_index()

# Create a line plot to answer the question.
sns.lineplot(x="appointment_month", y=0, hue ='appointment_mode', data=ar_hcp, ci=None)
```

**Question 5:** Are there any trends in time between booking an appointment?

```python
# Determine the total number of appointments per month.
ar_hcp = ar.groupby([ar['appointment_month'], ar['time_between_book_and_appointment']])[['time_between_book_and_appointment']].value_counts().reset_index()

# Create a line plot to answer the question.
sns.lineplot(x="appointment_month", y=0, hue ='time_between_book_and_appointment', data=ar_hcp, ci=None)
```
